## Create a density function using interpolation


- Write a density function density() that uses interpolation and takes in a random variable
- Use `np.histogram()`
- The function should return two lists carrying x and y coordinates for plotting the density function

```python
import numpy as np
import matplotlib.pyplot as plt

def density(x):
    """
    Compute the density (probability density function) of the given dataset x.
    
    Parameters:
    x (numpy array): Input dataset
    
    Returns:
    tuple: Tuple containing x values and corresponding density values.
    """
    # Calculate density using a histogram-like approach
    counts, bins = np.histogram(x, bins=30, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    return bin_centers, counts

# Example usage and test
np.random.seed(5)
mu, sigma = 0, 0.1  # mean and standard deviation
s = np.random.normal(mu, sigma, 100)
x, y = density(s)

plt.plot(x, y, label='Density')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Density Plot')
plt.legend()
plt.show()
```

## Add overlapping density plots to the histograms plotted earlier

```python
# Your code here 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example data
np.random.seed(5)
mu, sigma = 0, 0.1  # mean and standard deviation
s = np.random.normal(mu, sigma, 100)

# Plot histogram with density plot
plt.figure(figsize=(8, 6))

# Plot histogram with seaborn
sns.histplot(s, bins=30, kde=True, stat='density', color='skyblue', edgecolor='black', linewidth=1.2)

# Overlay density plot using seaborn
sns.kdeplot(s, color='red', linestyle='-', linewidth=2)

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with Overlaid Density Plot')
plt.legend(labels=['Density', 'Histogram'])

# Show plot
plt.show()
```
